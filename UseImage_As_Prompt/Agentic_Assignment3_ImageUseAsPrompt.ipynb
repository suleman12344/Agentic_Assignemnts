```python
import google.generativeai as genai
from google.colab import userdata
GEMINI_API_KEY: str = userdata.get('GOOGLE_API_KEY')
if GEMINI_API_KEY:
  print("API key found")
else:
  print("No API key found")
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

```python
dir(genai)
```

```python
for model in genai.list_models():
  print(model)
```

```python
from google.generativeai.generative_models import GenerativeModel
model: GenerativeModel = genai.GenerativeModel("gemini-1.5-flash")
```

```python
dir(model)
```

```python
from google.generativeai.types.generation_types import GenerateContentResponse
response: GenerateContentResponse = model.generate_content(
    "Explain the basics of LLM in simple terms. Respond in Markdown format."
)
```

```python
from IPython.display import display
from IPython.display import Markdown
display(Markdown(response.text))
```

## **USE IMAGES IN YOUR PROMPT**

```python
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
import PIL.Image
image=list(uploaded.keys())[0]
print(image)
```

```python
img = PIL.Image.open(image)
display(img)
```

```python
prompt = """This image contains Gender Discrimination and inequality along with some notes.
Given the describe the image as thoroughly as possible based on what you
see in the image, making sure to note all of the product features. Return output in markdown format:
"""
```

```python
response: GenerateContentResponse = model.generate_content([prompt, img])
display(Markdown(response.text))
```
